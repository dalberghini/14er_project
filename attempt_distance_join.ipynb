```python
import pandas as pd
from math import sin, cos, sqrt, atan2, radians
```

```python
joined = pd.read_csv("Resources/joined.csv")
camps = pd.read_csv("Resources/southwest_camp_data.csv")
mountains = pd.read_csv("Resources/14ers_data.csv", encoding = "ISO-8859â€“1")
```

```python
camp = camps[["Long","Lat", "name", "type","phone", "dates", "# sites", "Elevation","State"]]
co_camp = camp[camp["State"] == "CO"]
co_example = co_camp.head()
```

```python
distance_list = []
for k,l,m in zip(mountains["Mountain Peak"],mountains["Lat"],mountains["Long"]):
    lat1 = radians(l)
    lon1 = radians(m)
    for i, j in zip(co_example["Lat"],co_example["Long"]):
        lat2 = radians(i)
        lon2 = radians(j)
    R = 3958.8
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = (R * c)
    if distance <= 60:
        distance_list.append(k)
        distance_list.append(distance)
print(distance_list)
```
