```python
test=all_trail_roads1.head()
trails=['14er 1', '14er 2', '14er 3', '14er 4', '14er 5']
df_list=[]
for each_trail in trails: 
    sample_df=all_trail_roads1[['Range', 'Trailhead', 'Road Type', each_trail]]
    sample_df.columns=['Range', 'Trailhead', 'Road Type', 'Mountain Peak']
    df_list.append(sample_df)
combined_df=pd.concat(df_list) # assumes axis=0
combined_df=combined_df.drop_duplicates()
combined_df=combined_df.reset_index(drop=True)
combined_df = combined_df.dropna()
combined_df
```

```python
camp = camps[["Long","Lat", "name", "type","phone", "dates", "# sites", "Elevation","State"]]
co_camp = camp[camp["State"] == "CO"]
co_example = co_camp
```

```python
distance_list = []
mountain_list = []
camp_list = []
for l,m,n in zip(mountains["Mountain Peak"],mountains["Peak Latitude"],mountains["Peak Longitude"]):
    lat1 = radians(m)
    lon1 = radians(n)
    for i, j, k in zip(co_example["Lat"],co_example["Long"], co_example["name"]):
        lat2 = radians(i)
        lon2 = radians(j)
        camp_name = k
        R = 3958.8
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))
        distance = (R * c)
        if distance <= 20:
            mountain_list.append(l)
            distance_list.append(distance)
            camp_list.append(camp_name)
```

```python
distance_df = pd.DataFrame({
    "Mountains": mountain_list,
    "Distance from Campsite (miles)" : distance_list,
    "name": camp_list})
distance_df = distance_df.drop_duplicates()
distance_df
```

```python
#This process created 6630 instances of campgrounds within 60 miles of the 14ers in CO, below is the counts of how many each has, unless I did everything wrong.
distance_df.groupby("Mountains").count()
```
