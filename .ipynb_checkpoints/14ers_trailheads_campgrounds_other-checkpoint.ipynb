```python
df_easy_dirt_road=tables[3]
df_easy_dirt_road["Road Type"] = "2WD, easy dirt road"

df_easy_dirt_road=df_easy_dirt_road.rename(columns=({0: "Range", 1: "Trailhead", 2: "Mountains"}))
df_easy_dirt_road
```

```python
df_rough_dirt_road=tables[4]
df_rough_dirt_road["Road Type"] = "2WD, rough dirt road"
df_rough_dirt_road=df_rough_dirt_road.rename(columns=({0: "Range", 1: "Trailhead", 2: "Mountains"}))
df_rough_dirt_road
```

```python
df_easy_good_clearance_road=tables[5]
df_easy_good_clearance_road["Road Type"] = "Easy 4WD"
df_easy_good_clearance_road=df_easy_good_clearance_road.rename(columns=({0: "Range", 1: "Trailhead", 2: "Mountains"}))
df_easy_good_clearance_road
```

```python
df_moderate_4WD_road=tables[6]
df_moderate_4WD_road["Road Type"] = "Moderate 4WD"

df_moderate_4WD_road=df_moderate_4WD_road.rename(columns=({0: "Range", 1: "Trailhead", 2: "Mountains"}))
df_moderate_4WD_road
```

```python
df_rough_4WD_road=tables[7]
df_rough_4WD_road["Road Type"] = "Rough 4WD"
df_rough_4WD_road=df_rough_4WD_road.rename(columns=({0: "Range", 1: "Trailhead", 2: "Mountains"}))
df_rough_4WD_road
```

```python
df_difficult_4WD_road=tables[8]
df_difficult_4WD_road["Road Type"] = "Difficult 4WD"
df_difficult_4WD_road=df_difficult_4WD_road.rename(columns=({0: "Range", 1: "Trailhead", 2: "Mountains"}))
df_difficult_4WD_road
```

```python
all_roads_df=df_paved.append(df_easy_dirt_road)
#all_roads_df
#df_rough_dirt_road,df_easy_good_clearance_road
```

```python
#use append to add all 7 tables together
all_roads_df=all_roads_df.append((df_rough_dirt_road,df_easy_good_clearance_road,\
                                  df_moderate_4WD_road,df_rough_4WD_road, df_difficult_4WD_road))
all_roads_df
```

```python
#drop duplicates
all_roads_df=all_roads_df.drop_duplicates()
all_roads_df

```

```python
#reset index
all_roads_df1=all_roads_df.reset_index(drop=True)

all_roads_df1
```

```python
columnsplit=all_roads_df['Mountains'].str.split(",", expand=True)

# df = df.assign(City=columnsplit[0],State=columnsplit[1])
# df.head()


trails=columnsplit.rename(columns=({0: "14er 1", 1: "14er 2", 2: "14er 3", 3: "14er 4", 4: "14er 5"}))
trails=trails.reset_index(drop=True)
trails
```

```python


trails
```

```python
all_trail_roads=pd.merge(all_roads_df,trails, right_index=True, left_index=True)
#trails=pd.merge(all_roads_df,trails, on="Range")
all_trail_roads
```

```python
all_trail_roads.drop_duplicates()
```

```python
all_trail_roads1=all_trail_roads.drop(['Mountains'],axis=1)
#all_trail_roads1
all_trail_roads1=all_trail_roads1.reset_index(drop=True)
all_trail_roads1
```

```python
test=all_trail_roads1.head()
trails=['14er 1', '14er 2', '14er 3', '14er 4', '14er 5']
df_list=[]
for each_trail in trails: 
    sample_df=all_trail_roads1[['Range', 'Trailhead', 'Road Type', each_trail]]
    sample_df.columns=['Range', 'Trailhead', 'Road Type', 'Mountain Peak']
    # display(sample_df)
    df_list.append(sample_df)
combined_df=pd.concat(df_list) # assumes axis=0
combined_df=combined_df.drop_duplicates()
combined_df=combined_df.reset_index(drop=True)
combined_df = combined_df.dropna()
combined_df
```
